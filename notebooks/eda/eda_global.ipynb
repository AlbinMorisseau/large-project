## Pie chart showing the distribution of the number of data points in each dataset

```python
import os
import csv
import matplotlib.pyplot as plt

# Dossier contenant les datasets
dataset_folder = "../data/processed/dataset"

# Dictionnaire pour stocker le nombre de lignes par fichier
dataset_sizes = {}

# Parcours de tous les fichiers CSV
for filename in os.listdir(dataset_folder):
    if filename.endswith(".csv"):
        filepath = os.path.join(dataset_folder, filename)
        print(f"Comptage des lignes dans {filename}...")

        # Comptage efficace sans tout charger
        with open(filepath, 'r', encoding='utf-8', errors='ignore') as f:
            reader = csv.reader(f)
            next(reader, None)  # sauter l’en-tête si présent
            line_count = sum(1 for _ in reader)
            dataset_sizes[filename] = line_count

# Vérification
```

```python
print("\nTailles des datasets :")
for name, size in dataset_sizes.items():
    print(f"{name}: {size} lignes")

# Création du camembert
plt.figure(figsize=(8, 8))
plt.pie(
    dataset_sizes.values(),
    labels=dataset_sizes.keys(),
    autopct="%1.1f%%",
    startangle=90
)
plt.title("Data distribution")
plt.show()
```

```python
plt.figure(figsize=(9, 6))
wedges, texts, autotexts = plt.pie(
    dataset_sizes.values(),
    startangle=90,
    autopct=lambda p: f'{p:.1f}%' if p > 3 else '',  
)

# Légende propre à côté
plt.legend(
    wedges,
    [f"{name} ({count:,} lines)" for name, count in dataset_sizes.items()],
    title="Datasets",
    loc="center left",
    bbox_to_anchor=(1, 0.5)
)
plt.title("Data distribution", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# --- Data ---
names = list(dataset_sizes.keys())
sizes = list(dataset_sizes.values())

# --- Sort in descending order ---
sorted_indices = np.argsort(sizes)[::-1]
names = [names[i] for i in sorted_indices]
sizes = [sizes[i] for i in sorted_indices]

# --- Color palette (viridis) ---
colors = plt.cm.viridis(np.linspace(0, 1, len(sizes)))

# --- Create the plot ---
plt.figure(figsize=(10, 6))
bars = plt.barh(names, sizes, color=colors)

# Invert Y-axis so the largest dataset is on top
plt.gca().invert_yaxis()

# Logarithmic scale on X-axis
plt.xscale("log")

# Labels and title
plt.xlabel("Number of rows (log scale)", fontsize=10)
plt.title("Dataset distribution", fontsize=14)

plt.yticks(fontsize=7)

# Add value labels next to each bar
for bar, size in zip(bars, sizes):
    plt.text(
        size * 1.05,                 # slightly to the right of the bar
        bar.get_y() + bar.get_height() / 2,
        f"{size:,}",                 # format with commas
        va="center", ha="left", fontsize=7
    )

# Layout adjustment
plt.tight_layout()
plt.savefig("dataset_distribution.png", dpi=300, bbox_inches='tight')  # PNG haute résolution
plt.show()
```

## Reviews size

```python
import polars as pl
import os
import matplotlib.pyplot as plt
import pandas as pd

folder_path = "../data/processed/dataset_clean"
stats_list = []

for file in os.listdir(folder_path):
    if file.endswith(".csv"):
        file_path = os.path.join(folder_path, file)
        print(f"→ Processing of {file} ...")
        
        df = pl.read_csv(file_path)
        
        if "review" not in df.columns:
            print(f"No row 'review' found in {file}")
            continue
        
        df = df.with_columns(
            pl.col("review").map_elements(
                lambda x: len(str(x)) if x is not None else 0,
                return_dtype=pl.Int64
            ).alias("review_len")
        )
        
        summary = df.select([
            pl.mean("review_len").alias("mean"),
            pl.min("review_len").alias("min"),
            pl.max("review_len").alias("max"),
            pl.var("review_len").alias("variance"),
            pl.std("review_len").alias("std_dev"),
        ])
        
        stats = summary.to_dicts()[0]
        stats["dataset"] = file
        stats_list.append(stats)

if stats_list:
    stats_df = pd.DataFrame(stats_list).set_index("dataset")
    print("\n=== Global statistics ===")
    print(stats_df)

    plt.figure(figsize=(10, 6))
    plt.bar(stats_df.index, stats_df["mean"])
    plt.title("Average size of reviews per dataset")
    plt.xlabel("Dataset")
    plt.ylabel("Average size (characters)")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()
else:
    print("Error")
```

```python
plt.figure(figsize=(10, 6))
plt.errorbar(stats_df.index, stats_df["mean"], 
             yerr=stats_df["std_dev"], fmt='o', capsize=5, label='mean ± standard deviation')
#plt.scatter(stats_df.index, stats_df["min"], marker="_", color="red", label="Min")
#plt.scatter(stats_df.index, stats_df["max"], marker="_", color="green", label="Max")
plt.title("Statistical distribution of review sizes by dataset")
plt.xlabel("Dataset")
plt.ylabel("Size (Characters)")
plt.legend()
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
